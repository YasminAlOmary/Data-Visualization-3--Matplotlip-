## Data Visualization -Matplotlip 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

```

```python
x=np.arange(1,40,1)
y=np.linspace(2,41,len(x)) 
print(x)
print(y) 
```

```python
# %matplotlib inline :  to appear all content in chart 
%matplotlib inline  
# plt.plot(x,y): to draw relation plot between x and y
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,10)) 
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,5)) 
plt.plot(x,y)
# plt.title('my chart '): to added title into chart.
plt.title('my chart ')
```

```python
#plt.plot(x): to draw x separately 
plt.plot(x)
```

```python
#plt.plot(y): to draw x separately 
plt.plot(y)
```

```python
plt.plot(x)
plt.plot(y)
```

```python
#plt.subplot(2,1,1) : to draw in separate chart 
plt.subplot(2,1,1)
plt.plot(x)
#plt.subplot(2,1,2): to draw in separate chart
plt.subplot(2,1,2)
plt.plot(y)
```

```python
plt.figure(figsize=(4,4)) 
#plt.subplot(2,1,1) : to draw in separate chart 
plt.subplot(2,1,1)
plt.plot(x)
plt.title('X chart') 
#plt.subplot(2,1,2): to draw in separate chart
plt.subplot(2,1,2)
plt.plot(y)
plt.title('y chart')
```

```python
plt.figure(figsize=(4,4)) 
plt.subplot(3,1,1)
plt.plot(x)
plt.title('X chart') 

plt.subplot(3,1,2)
plt.plot(y)
plt.title('y chart')

plt.subplot(3,1,3)
sns.boxplot(x)
```

```python
# create dict
my_title={'family':'arial','color':'blue','size':20}

#create dict containing font type and color and size
my_axis={'family':'arial','color':'green','size':20}
plt.plot (x)
#  draw a chart and call fontdict to draw a chart with specific criteria
plt.title('x chart',fontdict=my_tite)
```

```python
my_title={'family':'arial','color':'blue','size':20}


my_axis={'family':'arial','color':'green','size':20}
plt.plot (x)
plt.title('x chart',fontdict=my_tite)

# to added label for x axis and calling fontdict.
plt.xlabel('sequense',fontdict=my_axis)
# to added label for y axis and calling fontdict.
plt.ylabel('Value',fontdict=my_axis)
```

```python
my_title={'family':'arial','color':'blue','size':20}

my_axis={'family':'calibri','color':'aqua','size':20}
plt.plot (x)
plt.title('x chart',fontdict=my_tite)

plt.xlabel('sequense',fontdict=my_axis)
plt.ylabel('Value',fontdict=my_axis)
```

```python
plt.figure(figsize=(5,5)) 
my_title={'family':'arial','color':'blue','size':20}

my_axis={'family':'calibri','color':'aqua','size':20}
plt.plot (x)
plt.title('x chart',fontdict=my_tite)

plt.xlabel('sequense',fontdict=my_axis)
plt.ylabel('Value',fontdict=my_axis)

#to added ticks range in x -axis chart.
plt.xticks(range (1,40))
# to added ticks range in y -axis chart.
plt.yticks(range (1,40))
```
